```python
#Code by Ethan Martinez
#Exploring Youtube Video Data

import pandas as pd
import matplotlib as plt

yt_csv = pd.read_csv("youtube_data/USvideos.csv")
ds_json = pd.read_json("youtube_data/datasets_4549_466349_US_category_id.json")

yt_csv["trending_date"] = pd.to_datetime(yt_csv["trending_date"], format="%y.%d.%m")
yt_csv["publish_time"] = pd.to_datetime(yt_csv["publish_time"], format="%y-%m-%dT%H:%M:%S.%fZ", exact=False)

#remove uneccessary columns and rows
del yt_csv["video_id"]
del yt_csv["thumbnail_link"]
yt_csv = yt_csv[yt_csv['ratings_disabled'] == 0]
yt_csv = yt_csv.set_index("title")

#create col for like to dislike ratio
def create_like_dislike_col(row):
    likes = row['likes']
    dislikes = row['dislikes']
    if (dislikes < 1):
        return likes
    else:
        return likes / dislikes
    
like_to_dislike = pd.Series({}, dtype='float64')
like_dislike_ratio = yt_csv.apply(create_like_dislike_col, axis=1)
yt_csv.insert(8, "like:dislike", like_dislike_ratio)

#create column for days between publish time and trending date
trend_pub = pd.Series({}, dtype='float64')
trend_pub = yt_csv.apply(lambda row : abs(row["trending_date"].day - row["publish_time"].day), axis=1)
yt_csv.insert(5, "days_btwn_trending_and_publish", trend_pub)
```

In the next cells we look at the average like to dislike ratio of videos depending on the number of days they took to trend after they were published.

```python
#make new dataframe of w/ index of days between and columns like:dislike, likes, dislikes
df = pd.DataFrame({})
df["days_btwn_trending_and_publish"] = yt_csv["days_btwn_trending_and_publish"]
df["like:dislike"] = yt_csv["like:dislike"]
df["likes"] = yt_csv["likes"]
df["dislikes"] = yt_csv["dislikes"]

#plot mean like:dislike ratio over time
#plot1 = df.groupby(['days_btwn_trending_and_publish']).mean()['like:dislike'].plot()

#there are outliers in the data set, so we plot median like:dislike ratio against days between trending and publish 
plot1 = df.groupby(['days_btwn_trending_and_publish']).median()['like:dislike'].plot(title='Median Like to Dislike Ratio Over Time')
plot1.set_ylabel('Like to Dislike Ratio')
plot1.set_xlabel('Number of Days Between Trending and Publish Dates')
plot1.legend(['Median Like:Dislike'])    
```

After the first day, the like to dislike ratio seems to spike. After this initial first few days, the like to dislike ratio stays consistent until it has a large drop on the last day. For me, two reasons come to mind for this drop. The first is that the novelty of the the videos may have subsided after being online for almost a month, causing more users to dislike the video/less users to like the video and making the ratio drop. Another reason could be that the types of videos that take 29-30 days to start trending are also more divicive, leading to a lower like to dislike ratio. I explore these in the next section:

Let's look at the breakdown of likes and dislikes over time.

```python
#plot average likes over time
#plot2 = df.groupby(['days_btwn_trending_and_publish']).mean()['likes'].plot()
plot2 = df.groupby(['days_btwn_trending_and_publish']).median()['likes'].plot(title='Median Likes Over Time')

plot2.set_ylabel('Likes')
plot2.set_xlabel('Number of Days Between Trending and Publish Dates')
plot2.legend(['Median Likes'])
```

```python
#plot average dislikes over time
#plot3 = df.groupby(['days_btwn_trending_and_publish']).mean()['dislikes'].plot()
plot3 = df.groupby(['days_btwn_trending_and_publish']).median()['dislikes'].plot(title='Median Dislikes Over Time')

plot3.set_ylabel('Dislikes')
plot3.set_xlabel('Number of Days Between Trending and Publish Dates')
plot3.legend(['Median Dislikes'])
```

There is an uptick in the average dislikes over time towards the end which could be why we see the drop in the like to dislike ratio. Now lets look at the relationship between the days it takes for a certain category to start trending and if the categories that take more days have a lower like to dislike ratio.

```python
df = pd.DataFrame({})
df["category_id"] = yt_csv["category_id"]
df["like:dislike"] = yt_csv["like:dislike"]
plot4 = df.groupby(['category_id']).mean().plot(title='Mean Like:Dislike Per Category', kind='bar')
plot4.set_xlabel('Category ID')
plot4.set_ylabel('Like:Dislike')
plot5 = df.groupby(['category_id']).median().plot(title='Median Like:Dislike Per Category ID', kind='bar')
plot5.set_xlabel('Category ID')
plot5.set_ylabel('Like:Dislike')
```

```python
df = pd.DataFrame({})
df["category_id"] = yt_csv["category_id"]
df["days_btwn_trending_and_publish"] = yt_csv["days_btwn_trending_and_publish"]
plot6 = df.groupby(['category_id']).mean().plot(title='Mean Days Between Publishing and Trending per Category', kind='bar')
plot6.set_xlabel('Category ID')
plot6.set_ylabel('Days Between Publishing and Trending')

plot7 = df.groupby(['category_id']).median().plot(title='Median Days Between Publishing and Trending per Category', kind='bar')
plot7.set_xlabel('Category ID')
plot7.set_ylabel('Days Between Publishing and Trending')
```

Although there are certain categories of videos that do tend to get a lower like to dislike ratio, there is no pattern for one category of videos taking longer to trend after publishing than others. Therefore, the lower like:dislike ratio on day 30 is due to the large uptick of dislikes compared to other days.

*Note: This larger uptick of dislikes on day 30 could be a coincidence. More data exploration is necessary to find this out.
